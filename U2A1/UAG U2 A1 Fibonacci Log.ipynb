```python
import time

def multiplicar_matrices(A, B):
    """Multiplicación manual de matrices 2x2."""
    return [
        [A[0][0] * B[0][0] + A[0][1] * B[1][0], A[0][0] * B[0][1] + A[0][1] * B[1][1]],
        [A[1][0] * B[0][0] + A[1][1] * B[1][0], A[1][0] * B[0][1] + A[1][1] * B[1][1]]
    ]

def potencia_matricial_dv(matriz, n):
    """Exponenciación rápida (Divide y Vencerás)."""
    if n == 1:
        return matriz
    
    # Divide y Vencerás: calculamos n/2 recursivamente
    mitad = potencia_matricial_dv(matriz, n // 2)
    resultado = multiplicar_matrices(mitad, mitad)
    
    # Si el exponente es impar, multiplicamos por la base una vez más
    if n % 2 != 0:
        resultado = multiplicar_matrices(resultado, matriz)
        
    return resultado

def calcular_fibonacci_log(n):
    """Función principal para el n-ésimo número de Fibonacci."""
    if n == 0: return 0
    if n == 1: return 1
    
    matriz_q = [[0, 1], [1, 1]]
    final_matriz = potencia_matricial_dv(matriz_q, n)
    
    # El valor de f(n) se encuentra en la posición [0][1]
    return final_matriz[0][1]

# --- Prueba de rendimiento ---
n_objetivo = 10000
inicio = time.perf_counter()
resultado = calcular_fibonacci_log(n_objetivo)
fin = time.perf_counter()

print(f"Fibonacci({n_objetivo}) calculado en: {fin - inicio:.8f} segundos")
# El resultado es un número inmenso, mostramos solo los primeros 10 dígitos
print(f"Primeros 10 dígitos: {str(resultado)[:10]}...")
```
