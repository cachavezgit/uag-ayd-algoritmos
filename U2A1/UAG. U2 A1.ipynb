```python
def fibonacci_lineal(n):
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    
    # Inicializamos los dos primeros términos
    a, b = 0, 1
    
    # Iteramos desde 2 hasta n
    for _ in range(2, n + 1):
        # El nuevo término es la suma de los dos anteriores
        a, b = b, a + b
        
    return b

# Ejemplo de uso
n = 10
print(f"El término {n} de Fibonacci es: {fibonacci_lineal(n)}")
```
