```python
def multiplicar_matrices(A, B):
    """Multiplica dos matrices de 2x2."""
    c11 = A[0][0] * B[0][0] + A[0][1] * B[1][0]
    c12 = A[0][0] * B[0][1] + A[0][1] * B[1][1]
    c21 = A[1][0] * B[0][0] + A[1][1] * B[1][0]
    c22 = A[1][0] * B[0][1] + A[1][1] * B[1][1]
    return [[c11, c12], [c21, c22]]

def potencia_matricial(matriz, n):
    """Calcula la potencia n de una matriz usando Divide y Vencerás."""
    # Caso base: la matriz identidad (equivalente al 1 en números)
    if n == 1:
        return matriz
    
    # Divide: calculamos la potencia para n/2
    temp = potencia_matricial(matriz, n // 2)
    
    # Vencer y Combinar: elevamos al cuadrado la mitad calculada
    res = multiplicar_matrices(temp, temp)
    
    # Si el exponente era impar, multiplicamos una vez más por la base
    if n % 2 != 0:
        res = multiplicar_matrices(res, matriz)
        
    return res

def fibonacci_fast(n):
    """Calcula el n-ésimo número de Fibonacci usando la forma matricial."""
    if n == 0: return 0
    if n == 1: return 1
    
    # Definimos la matriz Q de Fibonacci: [[0, 1], [1, 1]]
    matriz_q = [[0, 1], [1, 1]]
    
    # Elevamos la matriz a la potencia n
    resultado_matriz = potencia_matricial(matriz_q, n)
    
    # Según la teoría, f(n) está en la posición [0][1] o [1][0]
    return resultado_matriz[0][1]

# Ejemplo de uso
n = 10
print(f"El término {n} de Fibonacci es: {fibonacci_fast(n)}")
```
